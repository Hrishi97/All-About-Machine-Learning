## Decision Tree Regression - 

### Decision tree builds classification or regression models in the form of a tree structure. 

### It breaks down the dataset into smaller and smaller subets while at the same time an associated decision tree is incrementally developed.

### The final result is a tree with decision nodes and leaf nodes

### A decision node has two or more branches, each representing values for the attribute tested.

### A leaf node represents the final outcome of the decisions made through decision nodes.

#### Note - 

#### The core alogrithm for building decision trees called ID3, uses top-down or greedy search through the space of possible branches with not backtracking. 

#### ID3 alogrithm can also be used to build a decision tree, by replacing Information gain with Standard Deviation Reduction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('Position_Salaries.csv')

X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X,y)
```

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid),1)

plt.scatter(X,y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.show()

plt.scatter(X,y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.show()
```
