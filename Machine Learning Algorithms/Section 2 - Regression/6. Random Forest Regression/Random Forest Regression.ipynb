### Random Forest Regression - 

### Every decision tree has high variance, but when we combine all the decision trees together in parallel then the resultant variance becomes low as each decision tree gets perfectly trained on that particular sample data. 

### Hence, the output doesn't depend on one but multiple decision trees.

### In case of regression problem, the final output is the mean of all decision trees, whereas in classification problem, the final output is taken by using the majority voting classifier.



### Definition - A Random Forest is an ensemble technique capable of performing both regression and classification tasks with the use of multiple decision trees and a technique called Bootstrap and Aggregation, commonly known as bagging. The basic idea behind this is to combine multiple decision trees in determining the final output rather than relying on individual decision trees. 

### Random Forest has multiple decision trees as base learning models. We randomly perform row sampling and feature sampling from the dataset forming sample datasets for every model. This part is called Bootstrap.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('Position_Salaries.csv')

X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X,y)
```

```python
regressor.predict([[6.5]])
```

```python
X_test = np.arange(min(X),max(X), 0.01)
X_test = X_test.reshape(len(X_test), 1)

plt.scatter(X,y,color = "blue")
plt.plot(X_test,regressor.predict(X_test),color = "red")
plt.show()
```
