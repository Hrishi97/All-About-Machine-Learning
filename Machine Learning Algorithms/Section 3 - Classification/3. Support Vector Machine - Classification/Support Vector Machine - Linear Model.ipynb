### Support Vector Machine - 

### SVMs are supervised learning algorithms, used for both classification and regression. But generally, they are used in classification problems

### SVM represents different classes in a hyperplane in multidimensional space. The hyperplane will be generated in an iterative manner so that the error can be minimized.

### The goal of SVM is to divide the datasets into classes to find a maximum marginal hyperplane.

### Important parameters in SVM:

### 1. Support Vectors - Data points that are close to the hyperplane is called support vectors. Seperating line will be defined with the help of these data points

### 2. Hyperplane - It is a decision plane which divides the set of objects having different classes

### 3. Margin - It is a gap between two lines on the closet data points of different classes. It can be calculated as the perpendicular distance from the line to the support vectors. Large margin is considered as a good margin and small margin is considered as a bad margin.



### The main goal of SVM is to divide the datasets into classes to find a maximum marginal hyperplane (MMH) and it can be done in the following two steps −

### 1. First, SVM will generate hyperplanes iteratively that segregates the classes in best way.

### 2. Then, it will choose the hyperplane that separates the classes correctly.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
print(dataset)
```

```python
X = dataset.iloc[:,0:2].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0, cache_size = 200)
```

```python
classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,classifier.predict(X_test)))
```
