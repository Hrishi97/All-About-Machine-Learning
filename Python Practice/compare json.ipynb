```python
def sortItems(item):
    if(isinstance(item, list)):
        return sorted((sortItems(i)) for i in item)
    elif(isinstance(item, dict)):
        return sorted( (i, sortItems(j) ) for i,j in item.items() )
    else:
        return item

sortItems(x) == sortItems(y)
```

```python
def convertjson(elem):
    if isinstance(elem, dict):
        jsonmap = {}
        for i in elem:
            if isinstance(elem.get(i),list):
                convertedset=convertjson(elem.get(i))
                jsonmap[i] = convertedset
            elif isinstance(elem.get(i), dict):
                jsonmap[i] = elem.get(i)
            else:
                jsonmap[i] = elem.get(i)
        return sorted(jsonmap.items())
    elif isinstance(elem, list):
        subset = []
        for i in elem:
            if isinstance(i,dict):
                subdict = convertjson(i)
                subset.append(json.dumps(subdict, sort_keys=True))
        return sorted(subset)
    else:
        return elem
```
